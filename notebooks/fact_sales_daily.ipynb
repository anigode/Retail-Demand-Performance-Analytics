```python
import numpy as np
import pandas as pd

np.random.seed(42)
```

```python
df_products = pd.read_csv(
    "C:/Users/User/Desktop/dashboard/PROJECTS_NOT_ON_GITHUB/Retail/data/processed/dim_product.csv"
)
df_products.head()
```

```python
dates = pd.date_range(
    start="2023-01-01",
    end="2024-12-31",
    freq="D"
)
```

```python
sales_rows = []

for _, product in df_products.iterrows():
    
    base_demand = np.random.randint(1, 6)
    
    for date in dates:
        
        # Seasonality
        if date.month in [10, 11, 12]:   # festive season
            seasonal_factor = 1.3
        elif date.month in [6, 7]:       # slow season
            seasonal_factor = 0.8
        else:
            seasonal_factor = 1.0
        
        qty = int(
            max(
                0,
                np.random.normal(base_demand * seasonal_factor, 2)
            )
        )
        
        sales_rows.append({
            "date": date,
            "product_id": product["product_id"],
            "category": product["category"],
            "quantity_sold": qty,
            "revenue": qty * product["price"]
        })
```

```python
df_sales_daily = pd.DataFrame(sales_rows)

df_sales_daily.head()
```
