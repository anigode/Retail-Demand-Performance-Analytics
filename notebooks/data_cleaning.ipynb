```python
df_products["brand"] =df_products["brand"].fillna("Unknown")
```

```python
df_products = df_products.drop_duplicates(subset=["product_id"])
```

```python
df_products = df_products[
    (df_products["price"] > 0) & (df_products["price"] < 100000)
]
```

```python
df_users.info()
```

```python
df_users.isnull().sum()
```

```python
df_users = df_users.drop_duplicates(subset=["user_id"])
```

```python
df_users = df_users[
    (df_users["age"] >= 13) & (df_users["age"] <= 100)
]
```

```python
df_users["gender"] = df_users["gender"].str.lower().str.strip()
df_users["gender"] = df_users["gender"].replace({
    "m": "male",
    "f": "female"
})
```
