```python
import requests
import pandas as pd
```

```python
# API endpoints
products_url = "https://dummyjson.com/products"
carts_url = "https://dummyjson.com/carts"
users_url = "https://dummyjson.com/users"

# Fetch data
products = requests.get(products_url).json()['products']
carts = requests.get(carts_url).json()['carts']
users = requests.get(users_url).json()['users']

```

```python
# Convert to DataFrames
df_products = pd.DataFrame(products)
df_carts = pd.DataFrame(carts)
df_users = pd.DataFrame(users)
```

```python
df_products = df_products.rename(columns={
    "id": "product_id",
    "title": "product_name"
})
```

```python
df_products = df_products[
    [
        "product_id",
        "product_name",
        "category",
        "brand",
        "price",
        "rating",
        "discountPercentage"
    ]
]
```

```python
df_users = df_users.rename(columns={"id": "user_id"})

df_users = df_users[
    [
        "user_id",
        "age",
        "gender"
    ]
]
```

```python
df_carts = df_carts.rename(columns={
    "id": "cart_id",
    "userId": "user_id"
})
```
