```python
import numpy as np
import pandas as pd 
```

```python
fact_sales =pd.read_csv("C:/Users/User/Desktop/dashboard/PROJECTS_NOT_ON_GITHUB/Retail/data/processed/fact_sales_daily.csv", parse_dates=["date"])

dates = pd.DataFrame({
    "date": fact_sales["date"].unique()
}).sort_values("date")
```

```python
np.random.seed(42)

days= np.arange(len(dates))

dates["avg_temperature"] = (
    15
    + 10 *np.sin(2*np.pi*days/365)
    + np.random.normal(0,2,len(days))
).round(1)
```

```python
dates["min_temperature"] = (dates["avg_temperature"] - np.random.uniform(2, 5, len(dates))).round(1)
dates["max_temperature"] = (dates["avg_temperature"] + np.random.uniform(2, 5, len(dates))).round(1)
```

```python
def weather_condition(temp):
    if temp <= 5:
        return "Snow"
    elif temp <= 12:
        return "Rain"
    else:
        return "Clear"

dates["weather_condition"] = dates["avg_temperature"].apply(weather_condition)
```

```python
dates["is_cold_day"] = (dates["avg_temperature"] <= 10).astype(int)
```

```python
fact_weather_daily = dates.copy()
```

```python
fact_weather_daily.to_csv(
    "C:/Users/User/Desktop/dashboard/PROJECTS_NOT_ON_GITHUB/Retail/data/processed/fact_weather_daily.csv",
    index=False
)
```
