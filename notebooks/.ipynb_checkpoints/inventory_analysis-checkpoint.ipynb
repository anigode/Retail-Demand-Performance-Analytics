```python
import pandas as pd
import numpy as np
```

```python
df_products = pd.read_csv("C:/Users/User/Desktop/dashboard/PROJECTS_NOT_ON_GITHUB/Retail/data/processed/dim_product.csv")
df_carts = pd.read_csv("C:/Users/User/Desktop/dashboard/PROJECTS_NOT_ON_GITHUB/Retail/data/processed/carts_clean.csv")

```

```python
inventory_base = (
    df_products
    .assign(
        current_stock=lambda x: np.random.randint(20, 500, size=len(x))
    )
)
```

```python
inventory_base["avg_daily_sales"] = np.random.uniform(
    0.5, 10, size=len(inventory_base)
)
```

```python
inventory_base["forecasted_demand"] = (
    inventory_base["avg_daily_sales"] *
    np.random.uniform(0.8, 1.5, size=len(inventory_base))
)
```

```python
inventory_base["days_of_stock"] = (
    inventory_base["current_stock"] /
    inventory_base["avg_daily_sales"]
)
```

```python
inventory_base["overstock_qty"] = np.where(
    inventory_base["days_of_stock"] > 30,
    inventory_base["current_stock"] - (30 * inventory_base["avg_daily_sales"]),
    0
)
```

```python
inventory_base["inventory_risk"] = np.where(
    inventory_base["days_of_stock"] < 7, "Stockout Risk",
    np.where(
        inventory_base["days_of_stock"] > 30,
        "Overstock Risk",
        "Healthy"
    )
)
```

```python
inventory_base["risk_reason"] = np.where(
    inventory_base["inventory_risk"] == "Stockout Risk",
    "Low stock relative to sales velocity",
    np.where(
        inventory_base["inventory_risk"] == "Overstock Risk",
        "Excess stock relative to demand",
        "Balanced inventory position"
    )
)
```

```python
inventory_base["movement_category"] = pd.cut(
    inventory_base["avg_daily_sales"],
    bins=[0, 2, 6, float("inf")],
    labels=["Slow", "Medium", "Fast"]
)
```

```python
inventory_kpis = inventory_base[
    [
        "product_id",
        "product_name",
        "category",
        "price",
        "current_stock",
        "avg_daily_sales",
        "days_of_stock",
        "forecasted_demand",
        "overstock_qty",
        "movement_category",
        "inventory_risk",
        "risk_reason"
    ]
]
```
